```python
transaction_data3 = pd.merge(transaction_data2, customer_data)
```

```python
transaction_data3.rename(columns={'Income (USD/Month)': 'Income'}, inplace=True)
```

```python
transaction_data3
```

```python
repeat_customer1 = transaction_data3['Customer ID'].value_counts()
df = pd.DataFrame(repeat_customer1)
```

```python
df.index.names = ['Customer ID']
df.rename(columns={'Customer ID': 'Number of Trips'}, inplace=True)
```

```python
df
```

```python
bracket1 = df[df['Number of Trips'] >= 40]
bracket2 = df[(df['Number of Trips'] >= 10) & (df['Number of Trips'] < 40)]
bracket3 = df[(df['Number of Trips'] >= 2) & (df['Number of Trips'] < 10)]
bracket4 = df[df['Number of Trips'] == 1]
```

```python
bracket4
```

```python
top_customer_count = pd.merge(bracket1, transaction_data3, on='Customer ID', how='inner')
frequent_customer_count = pd.merge(bracket2, transaction_data3, on='Customer ID', how='inner')
occasional_customer_count = pd.merge(bracket3, transaction_data3, on='Customer ID', how='inner')
one_time_customer_count = pd.merge(bracket4, transaction_data3, on='Customer ID', how='inner')
```

```python
top_customer_count2 = pd.DataFrame(top_customer_count.groupby('Company')['Number of Trips'].count())
top_customer_count2 = top_customer_count2.rename(columns={top_customer_count2.columns[0]: 'Number of Trips'})
top_customer_count2['Company'] = top_customer_count2.index

frequent_customer_count2 = pd.DataFrame(frequent_customer_count.groupby('Company')['Number of Trips'].count())
frequent_customer_count2 = frequent_customer_count2.rename(columns={frequent_customer_count2.columns[0]: 'Number of Trips'})
frequent_customer_count2['Company'] = frequent_customer_count2.index

occasional_customer_count2 = pd.DataFrame(occasional_customer_count.groupby('Company')['Number of Trips'].count())
occasional_customer_count2 = occasional_customer_count2.rename(columns={occasional_customer_count2.columns[0]: 'Number of Trips'})
occasional_customer_count2['Company'] = occasional_customer_count2.index

one_time_customer_count2 = pd.DataFrame(one_time_customer_count.groupby('Company')['Number of Trips'].count())
one_time_customer_count2 = one_time_customer_count2.rename(columns={one_time_customer_count2.columns[0]: 'Number of Trips'})
one_time_customer_count2['Company'] = one_time_customer_count2.index
```

```python
top_customer_count2
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x='Company', y='Number of Trips', data = top_customer_count2, label = "count", color = "green")
sns.barplot(x='Company', y='Number of Trips', data = frequent_customer_count2, label = "count", color = "blue")
sns.barplot(x='Company', y='Number of Trips', data = occasional_customer_count2, label = "count", color = "yellow")
sns.barplot(x='Company', y='Number of Trips', data = one_time_customer_count2, label = "count", color = "red")
```

```python
pink_total_transactions = transaction_data3[transaction_data3['Company'] == 'Pink Cab']['Customer ID'].count().sum()
yellow_total_transactions = transaction_data3[transaction_data3['Company'] == 'Yellow Cab']['Customer ID'].count().sum()
```

```python
percentage_top_customer_pink = top_customer_count2['Number of Trips'][0] / pink_total_transactions
percentage_top_customer_yellow = top_customer_count2['Number of Trips'][1] / yellow_total_transactions

percentage_frequent_customer_pink = frequent_customer_count2['Number of Trips'][0] / pink_total_transactions
percentage_frequent_customer_yellow = frequent_customer_count2['Number of Trips'][1] / yellow_total_transactions

percentage_occasional_customer_pink = occasional_customer_count2['Number of Trips'][0] / pink_total_transactions
percentage_occasional_customer_yellow = occasional_customer_count2['Number of Trips'][1] / yellow_total_transactions

percentage_one_time_customer_pink = one_time_customer_count2['Number of Trips'][0] / pink_total_transactions
percentage_one_time_customer_yellow = one_time_customer_count2['Number of Trips'][1] / yellow_total_transactions
```

```python
data4 = {'Pink Cab':percentage_one_time_customer_pink,
       'Yellow Cab':percentage_one_time_customer_yellow}
data3 = {'Pink Cab':percentage_occasional_customer_pink+percentage_one_time_customer_pink,
       'Yellow Cab':percentage_occasional_customer_yellow+percentage_one_time_customer_yellow}
data2 = {'Pink Cab':percentage_frequent_customer_pink+percentage_occasional_customer_pink+percentage_one_time_customer_pink,
       'Yellow Cab':percentage_frequent_customer_yellow+percentage_occasional_customer_yellow+percentage_one_time_customer_yellow}
data1 = {'Pink Cab':percentage_top_customer_pink+percentage_frequent_customer_pink+percentage_occasional_customer_pink+percentage_one_time_customer_pink,
       'Yellow Cab':percentage_top_customer_yellow+percentage_frequent_customer_yellow+percentage_occasional_customer_yellow+percentage_one_time_customer_yellow}
```

```python
plt.figure(figsize=(9,6))
plt.bar(data1.keys(), data1.values(), color = 'orange', label = '40+ Rides')
plt.bar(data2.keys(), data2.values(), color = 'blue', label = '10-40 Rides')
plt.bar(data3.keys(), data3.values(), color = 'green', label = '5-10 Rides')
plt.bar(data4.keys(), data4.values(), color = 'red', label = '1 Ride')
plt.xlabel('Company')
plt.ylabel('Proportion')
plt.legend(bbox_to_anchor=(1,1), loc='upper left')
plt.title("Proportion of Regular vs Infrequent Customers per Company")
```
